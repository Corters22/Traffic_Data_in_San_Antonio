```python
import pandas as pd
import datetime as date
from datetime import datetime
```

```python
csv_read= pd.read_csv("US_Accidents_Dec20.csv.zip")
full_data_df = pd.DataFrame(csv_read)
```

```python
start_time = []
lat = []
lng = []
city = []
county = []
state = []
zipcode = []
temp = []
humidity = []
pressure=[]
visibility = []
wind_speed = []
precip = []
weather = []

for index, row in full_data_df.iterrows():
    if full_data_df.loc[index, "State"]=='TX':
        start_time.append(full_data_df.loc[index, "Start_Time"])
        lat.append(full_data_df.loc[index, "Start_Lat"])
        lng.append(full_data_df.loc[index, "Start_Lng"])
        city.append(full_data_df.loc[index, "City"])
        county.append(full_data_df.loc[index, "County"])
        state.append(full_data_df.loc[index, "State"])
        zipcode.append(full_data_df.loc[index, "Zipcode"])
        temp.append(full_data_df.loc[index, "Temperature(F)"])
        humidity.append(full_data_df.loc[index, "Humidity(%)"])
        pressure.append(full_data_df.loc[index, "Pressure(in)"])
        visibility.append(full_data_df.loc[index, "Visibility(mi)"])
        wind_speed.append(full_data_df.loc[index, "Wind_Speed(mph)"])
        precip.append(full_data_df.loc[index, "Precipitation(in)"])
        weather.append(full_data_df.loc[index, "Weather_Condition"])        
        
```

```python
texas_df = pd.DataFrame({'Time': start_time, 
                        'Latitude': lat,
                        'Longitude': lng, 
                        'City': city, 
                        'County': county, 
                        'State': state,
                        'Zipcode': zipcode,
                        'Temperature(F)': temp,
                        'Humidity(%)': humidity, 
                        'Pressure(in)': pressure,
                        'Visibility(mi)': visibility,
                        'Wind_Speed(mph)': wind_speed,
                        'Precipitation(in)': precip,
                        'Weather_Condition': weather})

texas_df.head()
```

```python
texas_df.dropna(inplace=True)
texas_df.head()
```

```python
texas_df["Year"] = texas_df["Time"].str.slice(0, 4).astype(int)
texas_df["Month"] = texas_df["Time"].str.slice(5, 7).astype(int)
texas_df["Day"] = texas_df["Time"].str.slice(8, 10).astype(int)
```

```python
cleaned_texas_df = texas_df.loc[texas_df["Year"]==2018]
cleaned_texas_df.reset_index(drop=True)
```

```python
cleaned_texas_df.to_csv("cleaned_2018_texas_accidents.csv")
```
